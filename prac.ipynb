```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import matplotlib.image as mpimg
```

```python
x= np.linspace(0,5,11)
y = x**2
plt.plot(x,y)
plt.xlabel("lol")
plt.ylabel("lil")
plt.title("lelele")
```

```python
plt.subplot(2,2,1)

plt.subplot(2,2,2)
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

fig = plt.figure(figsize=(2,3),dpi=100)
axis1 = fig.add_axes([0.2, 0.4, 0.6, 0.6])
axis1.plot(x, y)
axis2 = fig.add_axes([0.2,0.4,0.3,0.2])
axis2.plot(x,y)
plt.show()
plt.savefig("boxplot.png")
```

```python
plt.scatter(x,y)
plt.hist(x)
plt.boxplot(x)
```
