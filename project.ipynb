```python
df['Goal_Diff'] = df['Team_1_Goals'] - df['Team_2_Goals']
df['Total_Shots'] = df['Team_1_Shots'] + df['Team_2_Shots']
df['Rank_Diff'] = df['Team_1_FIFA_Rank'] - df['Team_2_FIFA_Rank']
df['Possession_Diff'] = df['Team_1_Possession'] - df['Team_2_Possession']


```

```python
def get_result(row):
    if row['Winner'] == row['Team_1']:
        return 1
    elif row['Winner'] == row['Team_2']:
        return 2
    else:
        return 0

df['Result'] = df.apply(get_result, axis=1)
```
