##  Finding Missing data

```python
Missing_data = {
    "Student_ID": [101, 102, 103, 104, 105, 106, 107],
    "Name": ["Aman", "Priya", "Rahul", "Sneha", "Karan", "Riya", "Arjun"],
    "Age": [19, 20, np.nan, 19, 21, np.nan, 20],
    "Gender": ["Male", "Female", "Male", np.nan, "Male", "Female", "Male"],
    "Marks": [85, np.nan, 78, 88, 95, 90, np.nan],
    "City": ["Jaipur", "Delhi", np.nan, "Pune", "Bengaluru", "Mumbai", "Delhi"]
}
index = [1,2,3,4,5,6,7]
df = pd.DataFrame(Missing_data,index=index)
df.drop(["Student_ID"],axis =1,inplace = True)
df.isna()
df.isna().sum() #count how many null value in each coloumn
df.isna().any() #check is there any coloumn with no null value
df
```

## Removing Missing Data

```python
df.dropna(thresh=3) #Row wise
df
```

## Filling the Missing Dataset

```python

values = {
    "Age" : 10 , "Gender" :"Female" , "Marks" : 99 , "City" : "Kolkata"
}
df.fillna(values,inplace = True) #coloumn wise
df
```

## Merging ,Joining and Concatenation

```python
# DataFrame 1: Student Details
df1 = pd.DataFrame({
    "Student_ID": [101, 102, 103, 104],
    "Name": ["Aman", "Priya", "Rahul", "Sneha"],
    "Age": [19, 20, 18, 21]
})

# DataFrame 2: Salary and Bonus Details
df2 = pd.DataFrame({
    "Student_ID": [102, 103, 105],
    "Salary": [50000, 60000, 55000],
    "Bonus": [5000, 7000, 6000]
})

pd.merge(df1,df2,on="Student_ID")
```

```python
pd.merge(df1,df2,on = "Student_ID",how = 'inner')
```

```python
pd.merge(df1,df2,on="Student_ID",how= "outer")
```

```python
pd.merge(df1,df2,on="Student_ID",how= "left")
```

```python
pd.merge(df1,df2,on='Student_ID',how="right")
```

## Concatenation of 2 Dataframe

```python
df1 = pd.DataFrame({
    "A": ["a1", "a2", "a3"],
    "B": ["b1", "b2", "b3"],
    "C": ["c1", "c2", "c3"]
})

df2 = pd.DataFrame({
    "A": ["a4", "a5", "a6"],
    "B": ["b4", "b5", "b6"],
    "C": ["c4", "c5", "c6"]
})
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

## Joining 2 dataFrames

```python
df1 = pd.DataFrame({
    "Name" :["Manan","Rahul","Shayam"]
},index = [1,2,3])

df2 = pd.DataFrame({
    "Score" : [45,55,34]
},index = [2,3,4])
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df1.join(df2,how = "inner")
```

```python
df1.join(df2,how= 'outer')
```
