```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y =  x**2
plt.plot(x,y)
plt.title("Speed X Distance")
plt.xlabel("Distance")
plt.ylabel("Speed")
```

## Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y**2)

plt.subplot(2,2,2)

plt.subplot(2,2,3)
plt.plot(y,x)

plt.subplot(2,2,4)
plt.plot(x,y)
```

## Object Oriented way


```python
## Figure----->>>
fig = plt.figure()
#kitna percent khiskana h [left se,bottom se,size of widhth,height]
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.2,.6,0.5])
axis2.plot(x,y)



## Figsize----->>>>
## Dpi increases the clearity
fig = plt.figure(figsize=(12,8),dpi=100)

axis1 = fig.add_axes([0.1,0.1,.4,.8])
axis1.plot(x,y)


axis2 = fig.add_axes([0.3,0.2,.3,0.3])
axis2.plot(x,y)


```

## Types of Plot :

```python
#scatterplot
plt.scatter(x,y)

#histogram
plt.hist(x)

#boxplot
plt.boxplot(x)
```

## Saveplot

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.3,0.5,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")

```
